```python
import numpy as np
import os
from scipy.misc import imread, imresize
import matplotlib.pyplot as plt
%matplotlib inline
print ("PACKAGE LOADING")
```

```python
cwd = os.getcwd()
print ("Current folder is %s" % (cwd))

def print_typeshape(img):
    print("Type is %s" % (type(img)))
    print("Shape is %s" % (img.shape,))
    
flash = imread(cwd + "/./lenna.bmp")
print_typeshape(flash)

plt.figure(0)
plt.imshow(flash)
plt.title("Original Image with imread")
plt.draw()
```

```python
flash2 = imread(cwd + "/./lenna.bmp")
print_typeshape(flash2)

plt.figure(0)
plt.imshow(flash2)
plt.title("Original Image with Imread.astype(np.float)")
plt.draw()
```

```python
flash_small = imresize(flash, [100,100,3])
print_typeshape(flash_small)

plt.figure(1)
plt.imshow(flash_small)
plt.title("Resize Image")
plt.draw()
```

```python
def rgb2gray(rgb):
    if len(rgb.shape) is 3 :
        return np.dot(rgb[..., :3], [0.299, 0.587, 0.114])
    else:
        print("Current Image If GRAY!")
        return rgb
    
flash_small_gray = rgb2gray(flash_small)
print_typeshape(flash_small)

plt.figure(2)
plt.imshow(flash_small_gray, cmap=plt.get_cmap("gray"))
plt.colorbar()
plt.draw()
```
